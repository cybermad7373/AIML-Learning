```python
bios['height_catogery'] = bios['height_cm'].apply(lambda x : 'Short' if x < 165 else ('Avgerage' if x < 185 else 'Tall'))
```

```python
bios.sample()
```

```python
def categorize_athelet(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70 : 
        return 'Lightweight'
    elif row['height_cm'] < 155 and row['weight_kg'] < 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athelet, axis = 1)
```
