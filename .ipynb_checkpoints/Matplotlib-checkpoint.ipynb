```python
import matplotlib.pyplot as plt
import numpy as np
```

## plot

```python
plt.plot([1,2,3,4,5])
plt.ylabel("ytemp")
plt.title("sdfsdF")
plt.xlabel("xtemp")
plt.show()
```

```python
x = [1,2,3,4]
m1 = [98,87,85,65]
m2 = [79,97,85,58]

plt.plot(x,m1,label = "mark1", color = 'r',linewidth = 1 , linestyle = "--", marker = '*')
plt.plot(x,m2,'b*--' ,label = "mark2",markersize = 10)

plt.title("Marks" , fontdict = {'fontname':'times new roman', 'fontweight':10 ,'color':'blue'})

plt.ylabel("m")
plt.xlabel("n")

plt.xticks( [1,2,3,4,5])
plt.yticks([50,60,70,80,90])
plt.legend()     # indicated the label
# plt.savefig("./data/temp.png")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20))
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0, 5, 0.2)
print(t)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0, 5, 0.2)

plt.plot(t, t, 'r--')
plt.show()

plt.plot( t, t**2, 'bs')
plt.show()

plt.plot( t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0, 5, 0.2)


plt.subplot(1,3,1)
plt.plot(t, t, 'r--')
plt.title("title 1")
plt.xlabel('x')
plt.ylabel('Y')

plt.subplot(1,3,2)
plt.plot( t, t**2, 'bs')
plt.title("title 2")

plt.subplot(1,3,3)
plt.plot( t, t**3, 'g^')
plt.title("title 3")

plt.suptitle("TITLEE")
plt.show()
```

```python
# axes

x = np.linspace(0,10,100)

fig, axes = plt.subplots(2,3)

axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title('Sin')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x,np.cos(x))
axes[0,1].set_title('Cos')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x,np.tan(x))
axes[0,2].set_title('Tan')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x,np.exp(x))
axes[1,0].set_title('Exp')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x,np.log(x))
axes[1,1].set_title('Log')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x,x)
axes[1,2].set_title('Linear')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')

axes[1,2].plot(x,np.sqrt(x))
axes[1,2].set_title('Linear')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')


plt.tight_layout()
```

## Scatter

```python
x = np.random.rand(1000) * 10
y = np.random.rand(1000) * 10

plt.scatter(x,y , marker = 'D', edgecolors = 'red', linewidths = 2, alpha = 0.3 , s = 5)
plt.legend(loc = 'lower right')
plt.grid(True)
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## Barplot

```python
cour = ['a','b','c','d','e','f']
std = [25,36,14,32,65,12]
plt.bar(cour, std , width = 0.5,align = 'edge', edgecolor = 'black', linewidth = 3, color = '#0f0', hatch = '*')
plt.show()
```

```python
hatch = ['/','*','o','\\','|','--']
clr = ['red','blue','yellow','green','orange','magenta']
bars = plt.bar(cour, std , width = 0.5,align = 'edge', linewidth = 3)
for bar, hatch,color in zip(bars,hatch,clr):
    bar.set_hatch(hatch)
    bar.set_color(color)
    bar.set_edgecolor('black')
plt.show()
```

## Histogram

```python
ages = [21,23,24,98,58,78,50,70]
plt.hist(ages, bins = 5, orientation = 'horizontal')
print(sorted(ages))
plt.show()
```

## pie

```python
cour = ['a','b','c','d','e','f']
std = [25,36,14,32,65,12]
explode = [0.3,0,0,0,0.1,0]
plt.pie(std,labels = cour, shadow = True, explode = explode, autopct = '%.2f%%')
plt.legend(loc = 'lower right')
plt.show()
```

## 3D plots

```python
ax = plt.axes(projection="3d")

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x,y,z)
plt.show()
```

```python
ax = plt.axes(projection="3d")

x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
```

```python
ax = plt.axes(projection="3d")

x = np.random.randint(1,10,5)
y = np.random.randint(1,10,5)
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = np.random.randint(1,10,5)


ax.bar3d(x,y,z,dx,dy,dz)
plt.show()
```

```python
ax = plt.axes(projection="3d")

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) 

ax.plot_surface(X,Y,Z, cmap = 'plasma')
plt.show()
```
